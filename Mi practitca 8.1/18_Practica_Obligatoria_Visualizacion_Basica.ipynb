#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
#1.2.1 population_total(columna poblacin)
top_cities = pd_cities.sort_values("population_total", ascending=False).head(5)
top_cities
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.bar(
    top_cities["city"],
    top_cities["population_total"],
    color=["blue", "gray", "olive", "brown"],
    alpha=0.7
)
```

```python
for i, value in enumerate(top_cities["population_total"]):
    ax.text(
        i,
        value + 100,
        f"{value:,}",
        ha="center",
        va="bottom"
    )

ax.set_title("Ciudades ms pobladas de California")
fig
```

```python
#1.2.2 Histograma
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(
    pd_cities["area_total_sq_mi"],
    bins=50,
    color="steelblue"
)
```

```python
ax.set_title(
    "Histograma del rea total",
    fontname="Bradley Hand ITC",
    fontsize=18,
   
)
ax.set_xlabel("Superficie (km$^2$)", fontname="Bradley Hand ITC", fontsize=14)
ax.set_ylabel("Nmero de ciudades", fontname="Bradley Hand ITC", fontsize=14)


fig
```

```python
#1.2.3 Scatter de agua
fig, ax = plt.subplots(figsize=(8,5))

ax.scatter(
    pd_cities["area_water_km2"],
    pd_cities["population_total"],
    color="steelblue"
)

ax.set_title(
    "rea de agua vs poblacin",
    fontname="Bradley Hand ITC",
    fontsize=18
)

ax.set_xlabel("rea de agua (km$^2$)", fontname="Bradley Hand ITC",fontsize=14)
ax.set_ylabel("Poblacin", fontname="Bradley Hand ITC",fontsize=14)

```

```python
#1.2.4 Boxplot de la altura de las ciudades
fig, ax = plt.subplots(figsize=(8,5))

ax.boxplot(pd_cities["elevation_m"])

ax.set_title(
    "Boxplot de la altura de las ciudades",
    fontname="Bradley Hand ITC",
    fontsize=18
)

ax.set_ylabel("Altura (m)", fontname="Bradley Hand ITC")

```
